```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing_df = pd.read_csv('data/housing.csv')
```

```python
housing_df.head()
```

```python
housing_df.columns = housing_df.columns.str.upper()
housing_df.head()
```

```python
housing_df.columns.to_list()
```

```python
housing_df.info()
```

```python
housing_df.describe()
```

```python
housing_df['MEDIAN_INCOME'] = housing_df['MEDIAN_INCOME']*10000
```

```python
housing_df.head()
```

```python
housing_df.info()
```

```python
housing_df.isna().sum()
```

```python
housing_df['TOTAL_BEDROOMS']=housing_df['TOTAL_BEDROOMS'].fillna(housing_df['TOTAL_BEDROOMS'].mean())
```

```python
housing_df.isna().sum()
```

```python
housing_df = housing_df.astype({'TOTAL_ROOMS':'int', 'HOUSING_MEDIAN_AGE':'int', 'POPULATION':'int', 'HOUSEHOLDS':'int', 'MEDIAN_INCOME':'int', 'MEDIAN_HOUSE_VALUE':'int', 'TOTAL_BEDROOMS':'int'})
```

```python
housing_df.info()
```

```python
housing_df['MEDIAN_HOUSE_VALUE'].plot.hist(title='House value')
# housing_df.plot(x = 'MEDIAN_HOUSE_VALUE', y = 'OCEAN_PROXIMITY', kind = 'hist')
```

```python
X = housing_df['MEDIAN_HOUSE_VALUE']
Y = housing_df['MEDIAN_INCOME']
plt.scatter(X, Y)
plt.xlabel('House Value')
plt.ylabel('Median Income')
```

```python
grouped_p = housing_df.groupby('HOUSING_MEDIAN_AGE')
```

```python
plt.bar(housing_df.HOUSING_MEDIAN_AGE.unique(), grouped_p.size())
plt.title('House Age')
```

```python
labels = housing_df['OCEAN_PROXIMITY'].unique()
colors = ['g', 'r', 'g', 'b', 'm']
plt.figure(figsize=(7, 7))
plt.pie(grouped_ocp.size(), colors=colors, labels=labels, autopct='%1.1f%%')
plt.title('OCEAN PROXIMITY')
```

```python
plt.figure(figsize=(10,8), dpi= 240)
sns.heatmap(housing_df.corr(), xticklabels=housing_df.corr().columns, 
            yticklabels=housing_df.corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.title('Correlation of Housing Data')
```
